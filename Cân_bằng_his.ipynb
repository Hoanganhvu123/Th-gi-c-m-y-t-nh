```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
img = cv2.imread('/content/5-tac-pham-piano-noi-tieng-cua-wa-mozart.jpg')
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap = 'gray')
```

```python
m,n = gray_img.shape
```

```python
size = gray_img.size
P = size
print(P)
```

```python
def can_bang_hist(img, new_level):
  g = []
  hg = []
  tg = []
  fg = []
  Fg = []
  count = 0
  G_max = np.max(gray_img) + 1
  for l in range(0,G_max-1):
    for i in range(m):
      for j in range(n):
        if gray_img[i][j] == l:
          g.append(l)

  G = list(set(g))
  for i in range(len(G)):
    hg.append(g.count(G[i]))

  tg.append(hg[0])
  for i in range(1,len(G)):
    tg.append(hg[i] + tg[i-1])

  for i in range(len(G)):
    Fg.append(max(0,round(tg[i]/(m*n/new_level) -1)))
  
  img_phan_doan_cat_nguong = np.zeros([m, n])
  
  for l in g:
    for i in range(m):
        for j in range(n):
            if g[l] in G : 
              g[l] = Fg[G.index(g[l])]
              img_phan_doan_cat_nguong[i,j] = g[l]
  return img_phan_doan_cat_nguong
```

```python
if __name__ == "__main__":
    # reading image in gray scale
    img_goc = cv2.imread("/content/5-tac-pham-piano-noi-tieng-cua-wa-mozart.jpg",0)

    # gi ham  phn on bng ct ngng ton cc da trn ngng
    img_phan_doan = can_bang_hist(img,1)

    # V v hin th nh gc, histogram nh gc v nh phn on
    fig2 = plt.figure(figsize=(16, 9))  # To vng v t l 16:9
    #To 4 vng v con
    (ax1, ax2), (ax3, ax4) = fig2.subplots(2, 2)
    # Hin th nh gc
    ax1.imshow(img_goc, cmap='gray')
    ax1.set_title('nh gc')
    ax1.axis('off')

    # Hin th histogram nh gc
    ax2.hist(img_goc.flatten(),bins=256)
    ax2.set_title('Hitogram nh gc')

    # Hin th nh phn on
    ax3.imshow(img_phan_doan, cmap='gray')
    ax3.set_title('nh phn on da vo tm ngng c bn')
    ax3.axis('off')

    # Hin th histogram nh phn on
    ax4.hist(img_phan_doan.flatten(), bins=256)
    ax4.set_title('Hitogram nh phn on')
    plt.show()
```
